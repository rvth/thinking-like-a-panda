```python
#Grouping rating, showing the mean for the reviews, price and year (numeric) columns 
ratings = books.groupby('Rating').mean()
ratings
```

```python
#Rounding to whole numbers by using astype, after we have done groupby 
ratings.round(decimals = 0).astype(int)
```

```python
#Can turn into int while doing groupby, rather than after
ratings = books.groupby('Rating').mean().astype(int)
ratings
```

```python
#GROUPING
```

```python
#Grouping by multi index - get an average rating for each author 
rating = books.groupby(['Author'])[['Rating']].mean()
rating
```

```python
#Rounding by 1 decimal place, after grouping
rating.round(decimals = 1)
```

```python
#Rounding by 1 decimal place, during grouping
rating = books.groupby(['Author'])[['Rating']].mean().round(decimals = 1)
rating
```

```python
#Sorting values by highest rating first 
rating.sort_values(by='Rating', ascending=False)
```

```python
#PIVOTS 
```

```python
#Author pivot showing how many genres they have books in, their average rating and total number of reviews 

author_pivot = pd.pivot_table(books, index=['Author'], values=['Rating', 'Reviews', 'Genre'], aggfunc={'Rating':'mean', 'Reviews':'sum', 'Genre':'count'})
author_pivot
```

```python
#Book pivot showing how many years they have made the list, their average rating and how many ratings they have had 

book_pivot = pd.pivot_table(books, index=['Name'], values=['Rating', 'Reviews', 'Year'], aggfunc={'Rating':'mean', 'Reviews':'sum', 'Year':'count'})
book_pivot
```
